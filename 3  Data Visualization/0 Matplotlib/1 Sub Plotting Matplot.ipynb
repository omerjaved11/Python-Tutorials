# Matplot Subplots
In this section we are going to explore subplots

### Subplots  / Subplotting
- It is a technique for creating multiple plots in form of rows or columns.
- we use method of matplot library *subplot*

```python
#we import matplot library 
import matplotlib.pyplot as plt
```

```python
fig,axarr = plt.subplots(2,1,figsize=(12,8))
#we asked for (2,1) mean (rows,columns) 
# plot gave us 2 rows and 1 column
```

```python
print(fig)
```

```python
print(axarr)
# if we print over object we get two elements in a list
# whos type is numpy array

print(type(axarr))
```

```python
fig,axarr = plt.subplots(2,1,figsize=(12,8))
```

When pandas generates a bar chart, behind the scenes here is what
it actually does:
1. Generate a new matplotlib Figure object.
12. Create a new matplotlib AxesSubplot object, and assign it to the Figure.
3. Use AxesSubplot methods to draw the information on the screen.
4. Return the result to the user.

```python
fig,axarr = plt.subplots(2,2,figsize=(12,8))
```

```python
print(axarr)
#now we get a list which contain two lists
# and each list have two elements which are our plots.
print(axarr.shape)
# if we check out shape of object it give us 2 rows 2 columns
# we can plot our figures into them by calling them by them index
```

```python
import pandas as pd
reviews = pd.read_csv("Data/winemag-data_first150k.csv",index_col=0)
```

```python
reviews.head()
```

```python
fig, axarr = plt.subplots(2,1,figsize=(12,9))
reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0])
reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1])
```

we can plot as many subplots as we want

```python
fig, axarr = plt.subplots(2,2,figsize=(12,9))
reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0][0])
reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1][0])
```

- we can plot as many plots as we want
- we can give them different colors
- we can set different fontsize
- we can set title of each
- we can set xlabel,ylables of each plot

```python
fig, axarr = plt.subplots(2, 2, figsize=(15, 10))
reviews['points'].value_counts().sort_index().plot.bar(
ax=axarr[0][0], fontsize=12, color='red'
)
axarr[0][0].set_title("Wine Scores", fontsize=18)
reviews['variety'].value_counts().head(20).plot.bar(
ax=axarr[1][0], fontsize=12, color='green'
)

reviews['province'].value_counts().head(20).plot.bar(
ax=axarr[1][1], fontsize=12, color='blue'
)

reviews['price'].value_counts().plot.hist(
ax=axarr[0][1], fontsize=12, color='yellow'
)
axarr[1][0].set_title("Wine Varieties", fontsize=18)
axarr[1][1].set_title("Wine Origins", fontsize=18)
axarr[0][1].set_title("Wine Prices", fontsize=18)

#Tune the subplot layout.
#subplots_adjust(left=None, bottom=None,
#    right=None, top=None,wspace=None, hspace=None)

plt.subplots_adjust(hspace=1,wspace=1)
import seaborn as sns
sns.despine(bottom=True)
```

## Why subplots
- when we need to compare two of more plots.
- It enable faceting
- **Faceting** : Faceting is the act of breaking data variables up across multiple subplots, and combining those subplots into a single figure.

So instead of one bar chart, we might have, say, four, arranged together in a grid.
The recommended way to perform faceting is to use the seaborn FacetGrid facility.
