## Object oriented method
Introduction to the Object Oriented Method

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

To begin we create a figure instance. Then we can add axes to that figure:


```python
fig = plt.figure() # we define a empty window using figure()
#we will the window with axes
# fig.add_axes takes list as input 
#list contains 4 elements ( left of axes , bottom of axes , width , height)
axes = fig.add_axes([0.1,0.1,0.5,0.8]) 
```

```python
fig1 = plt.figure() # we define a empty window using figure()
#we will the window with axes
# fig.add_axes takes list as input 
#list contains 4 elements ( left of axes , bottom of axes , width , height)
axes1 = fig1.add_axes([0.4,0.4,0.8,0.8]) 
axes1.plot(x,y) # now we have plot x and y on that canvas
axes1.set_xlabel('X Label')    #Set x Label
axes1.set_ylabel('Y Label')    #Set y Label
axes1.set_title('Title')       #Set Title of plot
```

```python
fig = plt.figure()
axes2 = fig.add_axes([0.1,0.1,0.8,0.8])
axes3 = fig.add_axes([0.2,0.5,0.4,0.3]) 
#([% away from left side,% above from bottom, size width , size height ])
```

```python
fig = plt.figure()
axes2 = fig.add_axes([0.1,0.1,0.8,0.8])
axes3 = fig.add_axes([0.45,0.2,0.4,0.3])
```

```python
fig = plt.figure()  #create a figure

axes2 = fig.add_axes([0.1,0.1,0.8,0.8])
axes3 = fig.add_axes([0.2,0.5,0.4,0.3]) 
#([% away from left side,% above from bottom, size width , size height ])

axes2.plot(x,y,'r')
axes3.plot(y,x,'b')

axes2.set_xlabel('outer X')
axes2.set_ylabel('outer Y')
axes3.set_xlabel('outer X')
axes3.set_ylabel('outer Y')


axes2.set_title('Outer Plot')
axes3.set_title('Inner Plot')
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2) #define row and columns of plots
#axes.plot(x,y)
```

```python
fig, axes = plt.subplots(3,3) 
#fixing over laps 
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
#axes is a list so we can plot in a loop
for current_ax in axes:
        current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
#plot using index
axes[0].plot(x,y)
axes[0].set_title('first')
axes[1].plot(y,x)

axes[1].set_title('Second')
plt.tight_layout()
```

### Figure Size and DPI
Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
#figsize = (width , height) in inches 
fig = plt.figure(figsize = (8,2),dpi=100) #dpi = pixels per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig , axes = plt.subplots(nrows=2,ncols=1,figsize = (8,2)) #dpi = pixels per inch

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

## Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#we can define a color for line using color parameter within plot
ax.plot(x,y,color='orange')
ax.plot(y,x,color='#FF8DF5') # RGB Hex code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#we can define a color for line using color parameter within plot
#we can define width / thickness of line using parameter linewidth= 1,2..
ax.plot(x,y,color='purple' , linewidth = 3)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#we can define a color for line using color parameter within plot

#we can define width / thickness of line using parameter lw= 1,2..

#control line transparency using alpha argument
ax.plot(x,y,color='purple' , lw = 7 , alpha = 0.1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#we can define a color for line using color parameter within plot

#we can define width / thickness of line using parameter lw= 1,2..

#we can define style of line like dash line dotted line 
# linestyle = '--' ,ls = '-.' ,ls = ':',ls = 'steps'  
ax.plot(x,y,color='purple' , lw = 7 ,linestyle=':')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#we can define a color for line using color parameter within plot
#we can define width / thickness of line using parameter lw= 1,2..
#we can define style of line like dash line dotted line 
# linestyle = '--' ,ls = '-.' ,ls = ':',ls = 'steps'  

#we can define marker on graph where x,y have corresponding values
#marker = 'o' ,marker = '+' ,marker = '*' ,marker = '1' 
#markers can have different size using parameter markersize

ax.plot(x,y,color='purple' , lw = 1 ,ls='--',marker='o',markersize = 10)


```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#we can define marker on graph where x,y have corresponding values
#marker = 'o' ,marker = '+' ,marker = '*' ,marker = '1' 
#markers can have different size using parameter markersize

#change marker face color using markerfacecolor='red'
#change marker face edge or marker edge thickness using markeredgewith=1
#change marker edge color using markeredgecolor = 'green'
ax.plot(x,y,color='purple' , lw = 1 ,ls='--',marker='o',markeredgecolor='green',markersize = 10,markeredgewidth=3, markerfacecolor='yellow')

```

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'black',lw=2,ls='--')

#we can set limit of x-axis and y-axis using set_xlim and set_ylim

ax.set_xlim([1,4])
ax.set_ylim([1,16])
```
