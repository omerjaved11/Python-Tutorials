## Pandas Styling of Plots

```python
import pandas as pd
%matplotlib inline
```

```python
reviews = pd.read_csv("Data/winemag-data_first150k.csv")
reviews.head()
```

```python
reviews['points'].value_counts().sort_index()
```

```python
reviews['points'].value_counts().sort_index().plot.bar()
```

```python
#adjust width and height of plot in inches
reviews['points'].value_counts().sort_index().plot.bar(figsize=(10,5))
```

```python
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(8,10),
    color='violet')
```

```python
#adjust size of scale numbers

reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12,5),
    color='red',
    fontsize=20
)
```

```python
#set title of plot
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(10,2),
    color='orange',
    fontsize=15,
    title="Ranking Given by Wine Magazine")
```

```python
# import matplotlib as plt
import seaborn as sns
```

```python
# title font size is not readable so we can adjust it by using 
# matplot fucntions
plot=reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(10,2),
    color='orange',
    fontsize=15)
plot.set_title("Ranking Given by Wine Magazine",fontsize=25)
```

```python
plot=reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(10,6),
    color='orange',
    fontsize=15)
plot.set_title("Ranking Given by Wine Magazine",fontsize=25)
plot.set_xlabel("Index")
plot.set_ylabel("Points")
```

```python
#remove bottom border
plot=reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(10,6),
    color='orange',
    fontsize=15)
plot.set_title("Ranking Given by Wine Magazine",fontsize=25)
plot.set_xlabel("Index",fontsize=25)
plot.set_ylabel("Points")
sns.despine(bottom=True)
```
